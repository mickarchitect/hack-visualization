### State Lat/Lon Reference Table
- [Source](https://inkplant.com/code/state-latitudes-longitudes)

```python
st_lat_lon = pd.read_csv("Data/state_lat_lon.csv")
print(st_lat_lon.shape)
print(st_lat_lon.head())
```

### Join to add the lat/lon to the st_yr df

```python
st_yr_lat_lon = st_yr.merge(st_lat_lon, left_on='STATE_ABBR', right_on='State_Abbr')
st_yr_lat_lon
```

```python
f2019 = st_yr_lat_lon['REPORTING_YEAR'] == 2019
st_yr_2019 = st_yr_lat_lon.loc[f2019]
st_yr_2019.to_csv("Data/st_yr_2019.csv")
```
